# 2. Probability Distributions

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from distributions import Beta
```

## 2.1.1 The beta distributions

```python
for i, [a, b] in enumerate([[0.1, 0.1], [1, 1], [2, 3], [8, 4]]):
    plt.subplot(2, 2, i + 1)
    beta = Beta(a, b)
    x = np.linspace(0, 1, 100)
    plt.plot(x, beta.predict_dist(x))
```

```python
beta = Beta(2, 2)
plt.subplot(2, 1, 1)
plt.plot(x, beta.predict_dist(x))
plt.annotate("prior", (0.1, 1.2))

beta.fit(n_ones=1, n_zeros=0)
plt.subplot(2, 1, 2)
plt.plot(x, beta.predict_dist(x))
plt.annotate("posterior", (0.1, 1.2))
```
